Import the dependencies

```python
import numpy as np
import pandas as pd
import difflib
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

Data Collection and Pre-processing

```python
#loading data from csv to pandas
movies_data = pd.read_csv('/content/movies.csv')
```

```python
#print first 5 rows of dataframe
movies_data.head()
```

```python
#find rows x columns
movies_data.shape
```

```python
#selecting the relevant features for recommendation
selected_features = ['genres','keywords','tagline','cast','director']
print(selected_features)
```

```python
#replace null values with null strings
for feature in selected_features:
  movies_data[feature] = movies_data[feature].fillna('')
```

```python
#combine all the 5 selected features
combined_features = movies_data['genres']+' '+movies_data['keywords']+' '+movies_data['tagline']+' '+movies_data['cast']+' '+movies_data['director']
```

```python
print(combined_features)
```

```python
#convert text data to feature vectors
vectorizer = TfidfVectorizer()
```

```python
feature_vectors = vectorizer.fit_transform(combined_features)
```

```python
print(feature_vectors)
```

Find similarities using Cosine Similarity

```python
#getting the similarities score using cosine similarities
similarity = cosine_similarity(feature_vectors)
```

```python
print(similarity)
```

```python
print(similarity.shape)
```

```python
import difflib
import ipywidgets as widgets
from IPython.display import display
import pandas as pd
```

```python
movies_data['title'] = movies_data['title'].str.strip()

# Convert movie titles to a lowercase list
list_of_all_titles = movies_data['title'].str.lower().tolist()
print(list_of_all_titles)
```

```python
# Create input field and search button
movie_name_widget = widgets.Text(placeholder="Enter the movie title")
search_button = widgets.Button(description="Search")

# Display the widgets
display(movie_name_widget, search_button)
```

```python
# Global variable to store the closest match
global_closest_match = None

# Function to search for the closest match
def search_movie(_):
    global global_closest_match  # Store the match globally

    movie_name = movie_name_widget.value.strip().lower()  # Convert to lowercase
    if movie_name:
        find_close_match = difflib.get_close_matches(movie_name, list_of_all_titles, n=1)

        if find_close_match:
            global_closest_match = find_close_match[0]  # Store closest match
            print(f"Closest match: {global_closest_match}")
        else:
            print("No close match found.")
    else:
        print("Please enter a movie title.")

# Attach event listener to button
search_button.on_click(search_movie)
```

```python
# Global variable to store movie index
global_index_of_movie = None

def find_movie_index():
    global global_closest_match, global_index_of_movie

    if global_closest_match is not None:
        filtered_data = movies_data[movies_data['title'].str.lower() == global_closest_match.lower()]

        if not filtered_data.empty:
            global_index_of_movie = filtered_data.index[0]  # Store index globally
            print(f"Index of '{global_closest_match}': {global_index_of_movie}")
        else:
            print(f"Movie '{global_closest_match}' not found in dataset after case conversion.")
    else:
        print("No valid movie match found. Run the search first.")

# Run this function manually after searching
find_movie_index()
```

```python
# Get similar movies based on index
def get_similar_movies():
    global global_index_of_movie  # Access the stored index

    if global_index_of_movie is not None:
        similarity_score = list(enumerate(similarity[global_index_of_movie]))
        print(f"Similarity scores for index {global_index_of_movie}: {similarity_score}")
    else:
        print("Movie index not found. Run the index search first.")

# Run this manually after finding the index
get_similar_movies()
```

```python
len(similarity_score)
```

```python
#sort movies based on their similarity score
sorted_similar_movies = sorted(similarity_score, key = lambda x:x[1], reverse = True)
print(sorted_similar_movies)
```

```python
#print the name of similar movies based on the index
print('Movies suggested for you : \n')

i = 1
for movie in sorted_similar_movies:
  index = movie[0]
  title_from_index = movies_data[movies_data.index==index]['title'].values[0]
  if(i<30):
    print(i,'.',title_from_index)
    i+=1
```

```python
import difflib
import ipywidgets as widgets
import pandas as pd
from IPython.display import display


# Convert column to list
list_of_all_titles = movies_data['title'].tolist()

# Create input field and search button
movie_name_widget = widgets.Text(placeholder="Enter the movie title")
search_button = widgets.Button(description="Search")

# Display widgets
display(movie_name_widget, search_button)

# Function to search for a movie and recommend similar ones
def on_search(button):
    movie_name = movie_name_widget.value.strip().lower()  # Get user input and normalize case

    if not movie_name:
        print("Please enter a movie title.")
        return

    # Find closest match
    closest_match = difflib.get_close_matches(movie_name, list_of_all_titles, n=1)

    if not closest_match:
        print(f"No close match found for '{movie_name}'. Try a different title.")
        return

    closest_match = closest_match[0]
    print(f"Closest match: {closest_match}")

    # Find index of the closest match
    filtered_data = movies_data[movies_data['title'].str.lower() == closest_match.lower()]

    if filtered_data.empty:
        print(f"Movie '{closest_match}' not found in dataset.")
        return

    index_of_movie = filtered_data.index[0]  # Get DataFrame index
    print(f"Index of '{closest_match}': {index_of_movie}")

    # Get similarity scores
    try:
        similarity_score = list(enumerate(similarity[index_of_movie]))  # Assuming 'similarity' is defined
    except NameError:
        print("Similarity matrix not found. Ensure the similarity matrix is generated.")
        return

    print(f"Similarity scores for '{closest_match}': {similarity_score[:5]}")  # Display first 5 scores for preview

    # Sort movies based on similarity score
    sorted_similar_movies = sorted(similarity_score, key=lambda x: x[1], reverse=True)

    # Print recommended movies
    print("\nMovies suggested for you:\n")
    i = 1
    for movie in sorted_similar_movies[1:31]:  # Skip the first one (it’s the same movie)
        index = movie[0]
        if index in movies_data.index:
            title_from_index = movies_data.loc[index, 'title']
            print(f"{i}. {title_from_index}")
            i += 1

# Link button click event
search_button.on_click(on_search)
```
